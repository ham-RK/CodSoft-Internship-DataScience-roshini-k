```python
# codsoft internship data science task 4 advertising dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ad=pd.read_csv("advertising.csv")
print("dataset is loaded")

#check correlation
sns.heatmap(ad.corr(),annot=True,cmap='Purples')
plt.title("correlation matrix")
plt.show()


features=ad[['TV','Radio','Newspaper']]
target_sales=ad['Sales']
#train test split
from sklearn.model_selection import train_test_split
f_train,f_test,y_train,y_test=train_test_split(features,target_sales,test_size=0.3,random_state=5)

#model training
from sklearn.linear_model import LinearRegression
sales_model=LinearRegression()
sales_model.fit(f_train,y_train)

#evaluate
preds=sales_model.predict(f_test)
from sklearn.metrics import mean_absolute_error,r2_score
print("MAE:",round(mean_absolute_error(y_test,preds),2))
print("R2 score:",round(r2_score(y_test,preds),2))
```
