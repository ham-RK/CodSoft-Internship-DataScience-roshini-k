```python
# codsoft internship data science task 3 iris dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris_df=pd.read_csv("IRIS.csv")
print("dataset is loaded")

#explore the data
sns.pairplot(iris_df,hue='species')
plt.suptitle("Feature pairing by species",y=1.02)
plt.show()

#prepare inputs
featuress=iris_df.drop('species',axis=1)
target=iris_df['species']

#train test split
from sklearn.model_selection import train_test_split
f_train,f_test,s_train,s_test=train_test_split(featuress,target,test_size=0.25,random_state=2)

#train the model
from sklearn.tree import DecisionTreeClassifier
iris_model=DecisionTreeClassifier()
iris_model.fit(f_train,s_train)

#prediction and accuracy
predictions=iris_model.predict(f_test)
from sklearn.metrics import accuracy_score
print("accuracy iris model:",round(accuracy_score(s_test,predictions)*100,2),"%")
```
